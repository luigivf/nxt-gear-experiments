```python
import numpy as np
import pandas as pd

import plotly.express as px
```

```python
def load_tag_data():
    print("Loading data...")
    df = pd.read_csv("./data/df_station_tag.csv", header=0, delimiter=",")

    return df

df_tag = load_tag_data()

df_tag.head()
```

```python
def preprocess_tag_data(df):
    print("Preprocessing tag data...")
    df_tag = df.copy()

    return df_tag

df_tag = preprocess_tag_data(df_tag)

df_tag.head()
```
